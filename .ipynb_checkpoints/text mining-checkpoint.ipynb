```python
# Read comments files into a list of list
comment_list = []

for url in urls:
    video_id = url[len(url)-11:len(url)]
    comment_file_path = "liziqi_top_10" + '/' + video_id + '_comment.txt'
    comment_file = os.path.join(os.getcwd(), comment_file_path)
    with open(comment_file, encoding='utf-8', mode = 'r') as f:
        lines = f.readlines()
        comment_list.append(lines)
```

## 2. Portion of languages

```python
# splitting the text lines into words
# translation table is a global variable
# mapping upper case to lower case and
# punctuation to spaces
translation_table = str.maketrans(string.punctuation+string.ascii_uppercase,
                                     " "*len(string.punctuation)+string.ascii_lowercase)
```

```python
languages_list = []
english_comments = []
spanish_comments = []
chinese_comments = []
chinese_tw_comments = []
for n in range(10):
    d = {}
    english_comment = []
    chinese_comment = []
    chinese_tw_comment = []
    spanish_comment = []
    print("start detecting file " + str(n))
    #comment_list[n].remove("\n")
    for comment in comment_list[n]:
        comment.replace("\n", "")
        if comment != "":
            comment = comment.translate(translation_table)
            try:
                result = detect(comment)
                if result == "en":
                    english_comment.append(comment)
                elif result == "es":
                    spanish_comment.append(comment)
                elif result == "zh-cn":
                    chinese_comment.append(comment)
                elif result == "zh-tw":
                    chinese_tw_comment.append(comment)
            except:
                result = "error"
                #print("Error in detecting this line from file "+ str(n) + " " + comment)
                # I don't know why but there are still many empty lines
            if result not in d:
                d[result] = 1
            elif result in d:
                d[result] += 1
    languages_list.append(d)# here I should convert it to dataframe for easier operation
    english_comments.append(english_comment)
    spanish_comments.append(spanish_comment)
    chinese_comments.append(chinese_comment)
    chinese_tw_comments.append(chinese_tw_comment)
```

## Word Frequency

```python
def count_frequency(word_list): 
      
    D = {}
      
    for new_word in word_list:
          
        if new_word in D:
            D[new_word] = D[new_word] + 1
              
        else:
            D[new_word] = 1
              
    return D
```
